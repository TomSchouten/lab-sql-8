```python
#How many distinct (different) actors' last names are there?

data= 'SELECT COUNT(DISTINCT last_name) AS "distinct_last_names" \
FROM actor'

unique_last_names = pd.read_sql_query(data, engine)

unique_last_names.head()
```

```python
#Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.
data = "SELECT *, DATE_FORMAT(rental_date, '%%M') AS 'month',  DATE_FORMAT(rental_date, '%%W') AS 'weekday',\
CASE WHEN DATE_FORMAT(rental_date, '%%W') IN ('Saturday', 'Sunday') then 'weekend' WHEN DATE_FORMAT(rental_date, '%%W') IN ('Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday') then 'workday'\
ELSE 'INCORRECT DATE'\
END AS 'day_type'\
FROM sakila.rental"

what_day = pd.read_sql_query(data, engine)

what_day.head()
```

```python
#Get all films with ARMAGEDDON in the title.

data= "SELECT *\
FROM   film WHERE  title like '%%ARMAGEDDON%%'"

arma_title = pd.read_sql_query(data, engine)

arma_title.head(20)
```

```python
#Get 10 the longest films.

data= 'SELECT   *\
FROM     film ORDER BY length DESC LIMIT    10'

longest_film_10 = pd.read_sql_query(data, engine)

longest_film_10.head(20)
```

```python
#How many films include Behind the Scenes content?
data= "SELECT *\
FROM   film WHERE  special_features LIKE   ('%%behind the scenes%%')" 

extra_content = pd.read_sql_query(data, engine)

extra_content.head(20)
```

```python
#Which kind of movies (rating) have a mean duration of more than two hours?

data = 'SELECT    rating, round(AVG(length),2) AS "mean_movie_length"\
FROM      film GROUP BY  rating HAVING    AVG(length) >120'

two_hours= pd.read_sql_query(data, engine)

two_hours.head(20)
```

```python
# Rank films by length (filter out the rows that have nulls or 0s in length column). 
# In your output, only select the columns title, length, and the rank.
data= 'SELECT   title, length, DENSE_RANK()\
OVER     (ORDER BY length DESC) as "rank"\
FROM     sakila.film WHERE length IS NOT null AND length > 0 ORDER BY length DESC'

rank = pd.read_sql_query(data, engine)

rank.head()
```
